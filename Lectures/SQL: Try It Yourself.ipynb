```python
# Find Books Borrowed during a Specific Year
# Write a query to list titles of books that were borrowed in the year 2022

%%sql
SELECT DISTINCT b.Title
FROM Books b
INNER JOIN Borrow br
ON b.BookID = br.BookID
WHERE YEAR(br.BorrowDate) = 2022;
```

```python
# Find Books with Ratings Above their Genre's Average

%%sql
WITH GenreAvg AS (
  SELECT Genre, AVG(Rating) AS AvgRating
  FROM BookRatings br
  INNER JOIN Books b ON br.BookID = b.BookID
  GROUP BY Genre
)

SELECT b.Title, br.Rating, g.AvgRating
FROM Books b
INNER JOIN BookRatings br ON b.BookID = br.BookID
INNER JOIN GenreAvg g ON b.Genre = g.Genre
WHERE br.Rating > g.AvgRating;
```

```python
# List Highest Rated Books in each Genre
# Modify the query to list the highest-rated book in each genre, using WITH for genre-specific maximum ratings

%%sql
WITH GenreMax AS (
  SELECT Genre, MAX(Rating) AS MaxRating
  FROM BookRatings br
  INNER JOIN Books b ON br.BookID = b.BookID
  GROUP BY Genre
)

SELECT b.Title, br.Rating, g.MaxRating
FROM Books b
INNER JOIN BookRatings br ON b.BookID = br.BookID
INNER JOIN GenreMax g ON b.Genre = g.Genre
WHERE br.Rating = g.MaxRating;
```
